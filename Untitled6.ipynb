```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('Canada.csv')
plt.xlabel("Year")
plt.ylabel("Income($)")
plt.scatter(df.Year,df.Income,color="r")
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Year']],df.Income)
```
